```python
from vpython import *
import math
import random
import sys


def k(P_i,P_j):
    return (P_i-P_j)

def q(V_tilde_i,V_j):
    return (V_tilde_i-V_j)

def d_star_2(V_tilde_i,V_j,P_i,P_j):
    if (q(V_tilde_i,V_j).mag ==0):
        result = k(P_i,P_j).mag2
    else:
        result = (k(P_i,P_j)-(k(P_i,P_j).dot(q(V_tilde_i,V_j)))/(q(V_tilde_i,V_j).mag2)*q(V_tilde_i,V_j)).mag2
    return result

def Energy(V_tilde_i,V_j,P_i,P_j,sigma):
    exponent = -d_star_2(V_tilde_i,V_j,P_i,P_j)/(2*sigma*sigma)
    return math.pow(math.e,exponent)

def w_d(P_i,P_r,sigma):
    exponent = -k(P_i,P_r).mag2/(2*sigma*sigma)
    return math.pow(math.e,exponent)

def w_phi(P_i,P_r,beta):
    cos_phi = P_i.dot(P_r)/(P_i.mag*P_r.mag)
    return math.pow(((1+cos_phi)/2),beta)

def w(P_i,P_r,sigma,beta):
    w1=w_d(P_i,P_r,sigma)
    w2=w_phi(P_i,P_r,beta)
    return w1*w2

def I_i(V_tilde_i,S_set,i):
    S_temp_set = S_set[:i]+S_set[i+1:]
    I_i_V_i = 0
#     sigma_const = const_sigma
#     beta_const = const_beta
    for s_i in S_temp_set:
        I_i_V_i += w(S_set[i][0],s_i[0],20,3)*Energy(V_tilde_i,s_i[1],S_set[i][0],s_i[1],20)
    print(I_i_V_i)
    return I_i_V_i

def S_i(V_tilde_i,u_i,lamda1):
    result = (lamda1*(u_i - V_tilde_i).mag2)
    return result

def D_i(V_tilde_i,S_set,G_set,i,lamda2):
    if(G_set[i]-S_set[i][0])==0:
        result = 0
    else:
        result = (G_set[i]-S_set[i][0]).dot(V_tilde_i)/((G_set[i]-S_set[i][0]).mag*((V_tilde_i).mag))
#     print("juli:%f",result)
    return -lamda2*result

def E_i(V_tilde_i,S_set,G_set,i,u_i,lamda1,lamda2):
    result = I_i(V_tilde_i,S_set,i)+S_i(V_tilde_i,u_i,lamda1)+D_i(V_tilde_i,S_set,G_set,i,lamda2)
#     print(result)
    return result

def min_E_i(V_tilde_i_Set,S_set,G_set,i,u_i,lamda1,lamda2):
    min_energy = sys.maxsize
    d_i = V_tilde_i_Set[0]
    for v in V_tilde_i_Set:
        temp = E_i(v,S_set,G_set,i,u_i,lamda1,lamda2)
        if(temp<min_energy):
            min_energy = temp
            d_i = v
    return d_i
    

lamda1 = 0
lamda2 = 0.001
const_sigma = 50
const_beta = 4

directionSet = [vector(1,0,0),vector(0,1,0),vector(0,-1,0),vector(-1,0,0)]
directionSet += [vector(-math.sqrt(2)/2,-math.sqrt(2)/2,0),vector(-math.sqrt(2)/2,math.sqrt(2)/2,0),vector(math.sqrt(2)/2,-math.sqrt(2)/2,0),vector(math.sqrt(2)/2,math.sqrt(2)/2,0)]
# obstacleSet = []

# for i in range(20):
#     temp = cylinder(pos=vector(int(random.uniform(-500, 500)),int(random.uniform(-400, 400)),0),axis=vector(0,0,30),radius= 20,color=color.red)
#     obstacleSet.append([temp.pos,temp.radius])
    
plane = box (pos=vector(0,0,0), length=1000, height=800, width=0.1, color=color.blue) 

start1 = sphere (pos=vector(10,0,0), radius=5, color=color.yellow)
start2 = sphere (pos=vector(400,0,0), radius=5, color=color.yellow)
start1.velocity = vector(2000,0,0)
start2.velocity = vector(2000,0,0)
S_set = [[start1.pos,start1.velocity],[start2.pos,start2.velocity]]
G_set = [vector(400,0,0),vector(10,0,0)]
dt = 0.0005
t = 0 
direction_Set = [vector(1,0,0),vector(-1,0,0)]
while(1):
    rate(50)
    s_i_V_tilde_i_Set = []
    f_d = []
    for s_i in S_set:
        v = s_i[1].mag
        temp = []
        for d in directionSet:
            temp.append(v*d)
        s_i_V_tilde_i_Set.append(temp)
#         print(s_i_V_tilde_i_Set)
        
    for i in range(0,len(s_i_V_tilde_i_Set)):
        f_d.append(min_E_i(s_i_V_tilde_i_Set[i],S_set,G_set,i,s_i_V_tilde_i_Set[i][0],lamda1,lamda2))
#     print(f_d)
    
    for i in range(0,len(S_set)):
        S_set[i][0] = S_set[i][0] + f_d[i]*dt
#     for i in range(0,len(S_set)):
    start1.pos = S_set[0][0]
    start2.pos = S_set[1][0]
    
    for i in range(0,len(S_set)):
        S_set[i][1] = f_d[i]
    t += dt
```
